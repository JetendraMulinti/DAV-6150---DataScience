# Data Preparation

```python
# Convert categorical variables that are represented as '0' or '1' to 'category' datatype
binary_categorical_columns = [
    'data_channel_is_lifestyle', 'data_channel_is_entertainment', 
    'data_channel_is_bus', 'data_channel_is_socmed', 
    'data_channel_is_tech', 'data_channel_is_world',
    'weekday_is_monday', 'weekday_is_tuesday', 'weekday_is_wednesday',
    'weekday_is_thursday', 'weekday_is_friday', 'weekday_is_saturday',
    'weekday_is_sunday', 'is_weekend'
]

for col in binary_categorical_columns:
    news_df[col] = news_df[col].astype('category')

# 'url' and 'timedelta' are non-predictive, these can be dropped. 
news_df = news_df.drop('url', axis=1)
news_df = news_df.drop('timedelta', axis=1)

def convert_columns_to_int(df, column_list):
    
    for column in column_list:
        # Only convert columns that exist in the dataframe
        if column in df.columns:
            # Convert column to int, handle missing values and convert to NaN
            df[column] = pd.to_numeric(df[column], errors='coerce').fillna(0).astype(int)
        else:
            print(f"Column {column} not found in the DataFrame.")
    return df

# List of columns you want to convert to integer
columns_to_convert = ['n_tokens_title', 'n_tokens_content',
                        'n_unique_tokens', 'n_non_stop_words', 'n_non_stop_unique_tokens',
                        'num_hrefs', 'num_self_hrefs', 'num_imgs', 'num_videos', 'num_keywords',
                        'shares']

# Apply the function to your dataframe
news_df = convert_columns_to_int(news_df, columns_to_convert)


# Confirm the changes have been applied correctly
news_df.dtypes
```

```python
# Calculate the median number of shares
median_shares = news_df['shares'].median()

# Define the boundaries for the categories
low_threshold = 0.5 * median_shares
high_threshold = 1.5 * median_shares

# Create the 'share_level' indicator variable
news_df['share_level'] = pd.cut(news_df['shares'],
                                bins=[0, low_threshold, high_threshold, np.inf],
                                labels=['low', 'medium', 'high'],
                                include_lowest=True)

# Remove the 'shares' column to prevent collinearity
news_df.drop('shares', axis=1, inplace=True)

# Verify the changes
news_df['share_level'].value_counts()
```

```python
# List of categorical columns to encode
categorical_columns = ['weekday_is_monday', 'weekday_is_tuesday', 'weekday_is_wednesday', 
                       'weekday_is_thursday', 'weekday_is_friday', 'weekday_is_saturday', 
                       'weekday_is_sunday', 'is_weekend', 'data_channel_is_lifestyle', 
                       'data_channel_is_entertainment', 'data_channel_is_bus', 
                       'data_channel_is_socmed', 'data_channel_is_tech', 'data_channel_is_world']

# Encoding categorical variables
news_df = pd.get_dummies(news_df, columns=categorical_columns, drop_first=True)
news_df.head(2)
```

```python
# For feature scaling, StandardScaler or MinMaxScaler can be used from sklearn
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Assuming we're using StandardScaler here
scaler = StandardScaler()

# Scale only numerical features, excluding the newly created 'share_level' which is categorical
numerical_cols = news_df.select_dtypes(include=['int64', 'float64']).columns.tolist()
news_df[numerical_cols] = scaler.fit_transform(news_df[numerical_cols])

# Confirm data types are correct, particularly for categorical variables
news_df.dtypes
```

```python
# Data integrity check # Assuming 'num_videos' should be non-negative
assert news_df['num_videos'].min() >= 0, "Negative values found in 'num_videos'"
```

# Neural Network Modeling

Feature Selection

```python


# Convert categorical target variable to numerical for feature selection
# This step is required because some feature selection methods require numerical inputs
target_numerical = news_df['share_level'].cat.codes

# SelectKBest with chi-square statistic
# This is used for classification and requires non-negative values, hence we'll apply it on positive features only
positive_features = news_df[numerical_cols].clip(lower=0)  # Ensuring values are non-negative
selector_chi2 = SelectKBest(score_func=chi2, k='all')  # Replace 'all' with the number of features you wish to select
selector_chi2.fit(positive_features, target_numerical)
chi2_scores = selector_chi2.scores_

# Mutual information
selector_mutual_info = SelectKBest(score_func=mutual_info_classif, k='all')  # Replace 'all' with the number of features you wish to select
selector_mutual_info.fit(news_df[numerical_cols], target_numerical)
mutual_info_scores = selector_mutual_info.scores_

# Creating a DataFrame to compare scores
feature_scores = pd.DataFrame({'Feature': numerical_cols, 
                               'Chi2 Score': chi2_scores, 
                               'Mutual Info Score': mutual_info_scores})
feature_scores.sort_values(by='Mutual Info Score', ascending=False, inplace=True)

feature_scores
```

Model 1: Focus on Keyword and LDA Topics
Features:
1. kw_max_avg (High Chi2 and Mutual Info scores)
2. kw_avg_avg (Highest Mutual Info score)
3. LDA_02 (High Chi2 and Mutual Info scores)
4. LDA_03 (Significant Mutual Info score)

Model 2: Focus on Self-References and Keyword Analysis
Features:
1. self_reference_min_shares (Good Chi2 and Mutual Info scores)
2. self_reference_avg_sharess (Good Chi2 and Mutual Info scores)
3. kw_min_avg (Good Chi2 and Mutual Info scores)
4. kw_avg_min (Moderate Chi2 and Mutual Info scores)

Model 3: Mixed Features with LDA and Rate Metrics
Features:
1. LDA_01 (Good Chi2 and Mutual Info scores)
2. LDA_04 (Moderate Chi2 and Mutual Info scores)
3. rate_positive_words (Moderate Mutual Info score)
4. rate_negative_words (Good Chi2 and Mutual Info scores)

Modelling

```python
# Assuming 'news_df' is your DataFrame
X1 = news_df[['kw_max_avg', 'kw_avg_avg', 'LDA_02', 'LDA_03']]
X2 = news_df[['self_reference_min_shares', 'self_reference_avg_sharess', 'LDA_01', 'LDA_04']]
X3 = news_df[['kw_min_avg', 'self_reference_max_shares', 'rate_positive_words', 'rate_negative_words']]
y = news_df['share_level'].cat.codes  # Convert categorical data to integer codes

# Split data for each model
X1_train, X1_test, y_train, y_test = train_test_split(X1, y, test_size=0.2, random_state=42)
X2_train, X2_test = train_test_split(X2, test_size=0.2, random_state=42)
X3_train, X3_test = train_test_split(X3, test_size=0.2, random_state=42)
```

```python
###  Model Definition

# Define Model 1
model1 = Sequential([
    Dense(32, input_shape=(4,), activation='relu'),
    Dense(16, activation='relu'),
    Dense(3, activation='softmax')
])
model1.compile(optimizer=Adam(learning_rate=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Define Model 2
model2 = Sequential([
    Dense(64, input_shape=(4,), activation='relu'),
    Dense(3, activation='softmax')
])
model2.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Define Model 3
model3 = Sequential([
    Dense(16, input_shape=(4,), activation='relu'),
    Dense(16, activation='relu'),
    Dense(16, activation='relu'),
    Dense(3, activation='softmax')
])
model3.compile(optimizer=Adam(learning_rate=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Training the models
history1 = model1.fit(X1_train, y_train, epochs=50, validation_split=0.2)
```

```python
history2 = model2.fit(X2_train, y_train, epochs=50, validation_split=0.2)
```

```python
history3 = model3.fit(X3_train, y_train, epochs=50, validation_split=0.2)
```

```python
### Evaluate each model

accuracy1 = model1.evaluate(X1_test, y_test)[1]
accuracy2 = model2.evaluate(X2_test, y_test)[1]
accuracy3 = model3.evaluate(X3_test, y_test)[1]
print(f'Model 1 Accuracy: {accuracy1}')
print(f'Model 2 Accuracy: {accuracy2}')
print(f'Model 3 Accuracy: {accuracy3}')
```

# Model Selection

```python
# Wrap Keras models in an estimator compatible with scikit-learn
estimator1 = KerasClassifier(build_fn=model1, epochs=50, batch_size=5, verbose=0)
estimator2 = KerasClassifier(build_fn=model2, epochs=50, batch_size=5, verbose=0)
estimator3 = KerasClassifier(build_fn=model3, epochs=50, batch_size=5, verbose=0)

# Perform k-fold cross-validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

results1 = cross_val_score(estimator1, X1_train, y_train, cv=kfold)
results2 = cross_val_score(estimator2, X2_train, y_train, cv=kfold)
results3 = cross_val_score(estimator3, X3_train, y_train, cv=kfold)

print(f'Model 1 Cross-Validation Accuracy: {results1.mean()}')
print(f'Model 2 Cross-Validation Accuracy: {results2.mean()}')
print(f'Model 3 Cross-Validation Accuracy: {results3.mean()}')
```

```python
# Assuming results1, results2, and results3 are the cross-validation results for models 1, 2, and 3 respectively
cv_accuracies = [results1.mean(), results2.mean(), results3.mean()]
models = ['Model 1', 'Model 2', 'Model 3']

# Find the index of the best model
best_model_index = cv_accuracies.index(max(cv_accuracies))
best_model_name = models[best_model_index]

print(f'{best_model_name} is the best model with a Cross-Validation Accuracy of: {cv_accuracies[best_model_index]}')
```
